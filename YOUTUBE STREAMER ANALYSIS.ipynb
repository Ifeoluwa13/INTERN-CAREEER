```python
import pandas as pd
from sklearn.model_selection import train_test_split

# Assuming 'Categories' is the column to be predicted (dependent variable)
# If 'Categories' is not the dependent variable, replace it with the actual target variable

# Create a copy of the original dataframe
youtubers_copy = youtubers_df.copy()

# Split the dataframe into train and test based on the presence of 'Categories'
train_df = youtubers_copy[youtubers_copy['Categories'].notnull()]
test_df = youtubers_copy[youtubers_copy['Categories'].isnull()]

# If 'Categories' is the dependent variable, drop it from the test set
test_df = test_df.drop('Categories', axis=1)

# Optionally, you can reset the index for both dataframes
train_df.reset_index(drop=True, inplace=True)
test_df.reset_index(drop=True, inplace=True)

# Display information about the two dataframes
print("Train DataFrame:")
print(train_df.info())

print("\nTest DataFrame:")
print(test_df.info())
```

```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import TfidfVectorizer

# Assuming you have a DataFrame named 'train_df' with columns: 'Username', 'Likes', 'Comments', 'Suscribers', 'Visits', 'Categories'
# Assuming each row represents a user's interaction with a streamer

# Aggregate interactions (Likes, Comments, etc.) for each user-streamer pair
user_streamer_agg = train_df.groupby(['Username', 'Categories']).agg({
    'Likes': 'sum',
    'Comments': 'sum',
    'Suscribers': 'mean',  # Use mean as an example; you may need to choose an appropriate aggregation
    'Visits': 'mean'
}).reset_index()

# Pivot table to get a user-streamer matrix
user_streamer_matrix = user_streamer_agg.pivot_table(index='Username', columns='Categories', values='Likes', fill_value=0)

# Collaborative Filtering (simplified implicit feedback approach)
user_similarity = cosine_similarity(user_streamer_matrix)
user_similarity_df = pd.DataFrame(user_similarity, index=user_streamer_matrix.index, columns=user_streamer_matrix.index)

def collaborative_recommendation(username, user_similarity_df, user_streamer_agg):
    user_ratings = user_streamer_matrix.loc[username]
    similar_users = user_similarity_df[username].sort_values(ascending=False)[1:6]
    similar_users_ratings = user_streamer_matrix.loc[similar_users.index]
    collaborative_recommendations = similar_users_ratings.mean(axis=0).sort_values(ascending=False)
    collaborative_recommendations = collaborative_recommendations[~collaborative_recommendations.index.isin(user_ratings.index)]  # Exclude already interacted items
    return collaborative_recommendations

# Example: Get collaborative recommendations for a user
username = 'MrBeast'
collaborative_recommendations = collaborative_recommendation(username, user_similarity_df, user_streamer_agg)

print("Collaborative Filtering Recommendations:")
print(collaborative_recommendations.head(10))
```

```python
# Content-Based Filtering
streamers_data = train_df[['Username', 'Categories', 'Suscribers', 'Visits', 'Likes', 'Comments']]
streamers_data = streamers_data.set_index('Username')

# Use TF-IDF to convert text (Categories) into numerical features
tfidf_vectorizer = TfidfVectorizer(stop_words='english')
categories_matrix = tfidf_vectorizer.fit_transform(streamers_data['Categories'])
categories_similarity = cosine_similarity(categories_matrix)
categories_similarity_df = pd.DataFrame(categories_similarity, index=streamers_data.index, columns=streamers_data.index)

def content_based_recommendation(username, categories_similarity_df, train_df):
    user_interactions = train_df[train_df['Username'] == username]
    user_interacted_streamers = user_interactions['Username'].unique()
    content_based_scores = categories_similarity_df[user_interacted_streamers].mean(axis=0).sort_values(ascending=False)
    content_based_recommendations = content_based_scores[~content_based_scores.index.isin(user_interacted_streamers)]  # Exclude already interacted items
    return content_based_recommendations

# Example: Get content-based recommendations for a user
content_based_recommendations = content_based_recommendation(username, categories_similarity_df, train_df)

print("\nContent-Based Filtering Recommendations:")
print(content_based_recommendations.head(10))
```

```python
def hybrid_recommendation(username, collaborative_recommendations, content_based_recommendations):
    hybrid_recommendations = 0.7 * collaborative_recommendations + 0.3 * content_based_recommendations
    return hybrid_recommendations

# Example: Get hybrid recommendations for a user
hybrid_recommendations = hybrid_recommendation(username, collaborative_recommendations, content_based_recommendations)

print("\nHybrid Recommendations:")
print(hybrid_recommendations.head(10))
```
